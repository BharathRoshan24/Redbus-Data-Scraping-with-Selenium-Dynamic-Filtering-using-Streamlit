```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_ksrtc_buses=pd.read_csv("ksrtc_buses.csv")
df_apsrtc_buses=pd.read_csv("apsrtc_buses.csv")
df_tgsrtc_buses=pd.read_csv("tgsrtc_buses.csv")
df_ktcl_buses=pd.read_csv("ktcl_buses.csv")
df_rsrtc_buses=pd.read_csv("rsrtc_buses.csv")
df_ctu_buses=pd.read_csv("ctu_buses.csv")
df_hrtc_buses=pd.read_csv("hrtc_buses.csv")
df_astc_buses=pd.read_csv("astc_buses.csv")
df_upsrtc_buses=pd.read_csv("upsrtc_buses.csv")
df_west_bengal_buses=pd.read_csv("west_bengal_buses.csv")


# Concatenate all DataFrames into a single one
Final_df = pd.concat([
    df_ksrtc_buses, 
    df_apsrtc_buses, 
    df_tgsrtc_buses, 
    df_ktcl_buses, 
    df_rsrtc_buses, 
    df_ctu_buses, 
    df_hrtc_buses, 
    df_astc_buses, 
    df_upsrtc_buses, 
    df_west_bengal_buses
], ignore_index=True)

Final_df
```

# Data cleaning

```python
# data about the data
Final_df.info()
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0) 
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True) 
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
# Save the DataFrame as a CSV
csv_path = "Final_busdetails_df.csv"
Final_df.to_csv(csv_path, index=False)

print(f"Final combined dataset saved as '{csv_path}'")
```

```python
cleaned_df = Final_df.dropna(subset=['Bus_name', 'Bus_type', 'Start_time', 'End_time', 
                                      'Total_duration', 'Price', 'Seats_Available', 
                                      'Ratings', 'Route_link', 'Route_name'])

# Convert cleaned DataFrame to a list of lists for insertion
data = cleaned_df.values.tolist()
```
